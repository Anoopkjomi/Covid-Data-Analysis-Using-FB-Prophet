**Renaming the columns Province/Country into Province and Country/Region into Country**

```python
def Import_Covid19data(filepath):
    data = filepath
    data = pd.read_csv(data)
    data.rename(columns={'Province/State':'Province','Country/Region':'Country'},inplace = True)
    return data
```

**Transforming given data by adding a column for dates**

```python
def transform_Covid19data(data):
    data = pd.melt(
    frame = data,
    id_vars = ['Province','Country','Lat','Long','Status'],
    var_name = 'Date',
    value_name = 'Count'
    )
    return data
```

```python
ConfirmedCases=Import_Covid19data("/content/time_series_covid19_confirmed_global.csv")
ConfirmedCases
```

*adding a column named Status and gives the value confirmed*

```python
ConfirmedCases['Status'] = 'Confirmed'
```

```python
ConfirmedCases = transform_Covid19data(ConfirmedCases)
```

**selecting top 15 countries that has highest number of confirmed,recovered,and death covid cases**

```python
def Top15CountriesCases(Status):
    Top15Countries = data[(data['Status']==Status) & (data.index==max(data[data['Status']==Status].index))].sort_values(by='Count',ascending = False).head(15)[['Country','Count']]
    return Top15Countries
```

```python
Top15CountriesRecovered = Top15CountriesCases('Recovered')
Top15CountriesConfirmed = Top15CountriesCases('Confirmed')
Top15CountriesDeaths = Top15CountriesCases('Death')
```

```python
def Display_Top15Country(Top15Countries,Status):
    Top15Countries = Top15Countries.reset_index()
    Top15Countries = Top15Countries[['Country','Count']]
    date = max(data[data['Status']==Status].index)
    date = date.date()
    print(Status +' cases till:',date)
    return Top15Countries
```

```python
Top15CountriesConfirmed = Display_Top15Country(Top15CountriesConfirmed,'Confirmed')
Top15CountriesConfirmed.style.background_gradient(cmap = 'Reds')
                                                  
```

```python
Top15CountriesRecovered = Display_Top15Country(Top15CountriesRecovered,'Recovered')
Top15CountriesRecovered.style.background_gradient(cmap = 'Blues')
```

```python
Top15CountriesDeaths = Display_Top15Country(Top15CountriesDeaths,'Death')
Top15CountriesDeaths.style.background_gradient(cmap = 'Greens')
```

**Time Series Forecasting using FB Prophet:**

*It is a procedure for forecasting time series data based on an additive model where non-linear trends are fit with yearly, weekly, and daily seasonality. It works with time series data that have strong seasonal effects and several seasons of historical data*

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
from fbprophet import Prophet
```

```python
Confirmed = tranformeddata.groupby("Date").sum()['Confirmed'].reset_index()
Death = tranformeddata.groupby("Date").sum()['Death'].reset_index()
Recovered = tranformeddata.groupby("Date").sum()['Recovered'].reset_index()
```

```python
Confirmed
```

*Changing column names as ds and y.ds is the date and y is the count*

```python
Confirmed.columns = ["ds", "y"]
Confirmed["ds"] = pd.to_datetime(Confirmed["ds"])
```

```python
Confirmed.tail()
```

**creating future dates of 100 days**

```python
model = Prophet()
model.fit(Confirmed)
future = model.make_future_dataframe(periods=100)
future.tail()
# it is taking weekly seasonality
```

```python
forecast = model.predict(future)
```

```python
forecast.head()
```

```python
forecast[["ds","yhat","yhat_lower", "yhat_upper"]].tail()
```

*plotting the predicted projection* 

```python
Confirmed_forecast_plot = model.plot(forecast)
```

*The inference is the dotted points are actual data and the line is predicted data and the upper line is the trend.*
*based on the trend confirmed cases increases.*

*Visualize each components(trends,monthly and weekly)*

```python
confirmed_forecast_plot = model.plot_components(forecast)
```

**Hyper Parametric tunning**

```python

from fbprophet.diagnostics import cross_validation
confirmeddata = cross_validation(model, initial='130days', period='22days', horizon = '44days')
confirmeddata.head()
```

```python
from fbprophet.diagnostics import performance_metrics
confirmeddata_p = performance_metrics(confirmeddata)
confirmeddata_p.head()
```

```python
from fbprophet.plot import plot_cross_validation_metric
fig = plot_cross_validation_metric(confirmeddata, metric='rmse')
```
